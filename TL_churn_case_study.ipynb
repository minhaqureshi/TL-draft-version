### 2. Data Cleaning

```python
## Fetch Empty columns
trainDF.isnull().sum()
```

None of the row is completely null , none of the column is completely empty, so there is no need to drop any row or column

```python
### Find the Columns where 50% values are null
missing_per = trainDF.isnull().sum() * 100 / len(trainDF)
missing_value_trainDF = pd.DataFrame({'column': trainDF.columns,'missing%':missing_per})
missing_value_trainDF = missing_value_trainDF[missing_value_trainDF['missing%']>50]
missing_value_trainDF.sort_values('missing%',inplace = True,ascending=False)
missing_value_trainDF
```
